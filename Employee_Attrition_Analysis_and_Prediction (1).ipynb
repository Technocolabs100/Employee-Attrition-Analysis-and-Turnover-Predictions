```python
data["Attrition"]=data["Attrition"].replace({"No":0,"Yes":1 })
```

```python
data["OverTime"]=data["OverTime"].map({"No":0,"Yes":1})
data["Gender"]=data["Gender"].map({"Male":0,"Female":1})
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
encoding_columns=["BusinessTravel","Department","EducationField","JobRole","MaritalStatus"]
label_encoders={}
for column in encoding_columns:
  label_encoders[column]=LabelEncoder()
  data[column] = label_encoders[column].fit_transform(data[column])
```

```python
data.head()
```

```python
data["Over18"]=data["Over18"].replace({"Y":1,"N":0})
```

```python
data.head()
```

```python
data.info()
```

```python
x=data.drop(["Attrition"], axis=1)
y=data["Attrition"].values
```

```python
from collections import Counter
from imblearn.over_sampling import RandomOverSampler
print(Counter(y))
rus = RandomOverSampler(random_state=42)
x_over,y_over = rus.fit_resample(x,y)
print(Counter(y_over))
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train, y_test = train_test_split(x_over,y_over,test_size=0.2, random_state=42)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, roc_curve,roc_auc_score
```

```python
logreg=LogisticRegression()
logreg.fit(x_train,y_train)
```

```python
prediction=logreg.predict(x_test)
confusion_matrix=confusion_matrix(y_test,prediction)
print("Accuracy Score ",accuracy_score(y_test, prediction))
```

```python
conf_matrix = confusion_matrix(y_test, logreg.predict(x_test))

# Create the figure
fig = plt.figure(figsize=(8, 4))

# Add the first subplot for the confusion matrix heatmap
ax1 = fig.add_subplot(1, 2, 1)
sns.heatmap(pd.DataFrame(conf_matrix), annot=True, cmap="Blues", fmt="d", ax=ax1)
bottom, top = ax1.get_ylim()
ax1.set_ylim(bottom + 0.5, top - 0.5)
ax1.set_xlabel("Prediction")
ax1.set_ylabel("Expectation")

# Add the second subplot for the ROC curve
ax2 = fig.add_subplot(1, 2, 2)
pred_probability = logreg.predict_proba(x_test)[:, 1]
fpr, tpr, _ = roc_curve(y_test, pred_probability)
auc = roc_auc_score(y_test, pred_probability)
ax2.plot(fpr, tpr, label="Data 1, AUC = " + str(auc))
ax2.set_xlabel("False Positive Rate")
ax2.set_ylabel("True Positive Rate")
ax2.legend(loc=4)

# Display the plots
plt.tight_layout()
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, roc_curve, roc_auc_score
```

```python
dt = DecisionTreeClassifier(random_state=42)
dt.fit(x_train, y_train)
```

```python
y_pred = dt.predict(x_test)
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, dt.predict_proba(x_test)[:, 1])
fpr, tpr, thresholds = roc_curve(y_test, dt.predict_proba(x_test)[:, 1])


print("\nAccuracy Score:", accuracy)
print("\nROC AUC Score:", roc_auc)
print("\nROC Curve:")
print("False Positive Rate:", fpr)
print("True Positive Rate:", tpr)
```

```python
conf_matrix_dt = confusion_matrix(y_test, dt.predict(x_test))


fig = plt.figure(figsize=(16, 8))

ax1 = fig.add_subplot(1, 2, 1)
sns.heatmap(pd.DataFrame(conf_matrix_dt), annot=True, cmap="Blues", fmt="d", ax=ax1)
bottom, top = ax1.get_ylim()
ax1.set_ylim(bottom + 0.5, top - 0.5)
ax1.set_xlabel("Prediction")
ax1.set_ylabel("Expectation")
ax1.set_title("Decision Tree - Confusion Matrix")

ax2 = fig.add_subplot(1, 2, 2)
pred_probability_dt = dt.predict_proba(x_test)[:, 1]
fpr_dt, tpr_dt, _ = roc_curve(y_test, pred_probability_dt)
auc_dt = roc_auc_score(y_test, pred_probability_dt)
ax2.plot(fpr_dt, tpr_dt, label=f'Data 1, AUC = {auc_dt:.2f}')
ax2.set_xlabel("False Positive Rate")
ax2.set_ylabel("True Positive Rate")
ax2.set_title("Decision Tree - ROC Curve")
ax2.legend(loc=4)

plt.tight_layout()
plt.show()
```
