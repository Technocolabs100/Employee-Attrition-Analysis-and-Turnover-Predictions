# Feature selection

#### Categorical feature

```python
categorical_feat = ["BusinessTravel","Department","Education","EducationField","EnvironmentSatisfaction","Gender","JobInvolvement","JobLevel","JobRole","JobSatisfaction","MaritalStatus","OverTime","PerformanceRating","RelationshipSatisfaction","StockOptionLevel","WorkLifeBalance"]
```

```python
stat = []
```

```python
for var in categorical_feat:
    contingency_table = pd.crosstab(df_cleaned[var],df_cleaned["Attrition"])
    chi2 = stats.chi2_contingency(contingency_table)[0]
    p_value = stats.chi2_contingency(contingency_table)[1]
    num_rows, num_cols = contingency_table.shape
    v_cramer = np.sqrt(chi2/(df_cleaned.shape[0]*min(num_rows-1,num_cols-1)))
    stat.append([var,chi2,p_value,v_cramer])
```

```python
stat_df = pd.DataFrame(stat, columns = ["Variable","Chi2_square","P_value","V_Cramer"])
```

```python
stat_df.sort_values("P_value", ascending = True, inplace = True)
```

```python
stat_df
```

```python
cat_feature = ["OverTime","JobRole","JobLevel","StockOptionLevel","MaritalStatus","JobInvolvement","BusinessTravel","EnvironmentSatisfaction"]
```

#### Numerical feature

```python
numerical_feat = ["Age","DailyRate","DistanceFromHome","HourlyRate","MonthlyIncome","MonthlyRate","NumCompaniesWorked","PercentSalaryHike","TotalWorkingYears","TrainingTimesLastYear","YearsAtCompany","YearsInCurrentRole","YearsSinceLastPromotion","YearsWithCurrManager"]
```

```python
stat_2 = []
```

```python
for var in numerical_feat:
    U_stat, p_value1 = stats.mannwhitneyu(df_cleaned[df_cleaned['Attrition'] == 'Yes'][var], df_cleaned[df_cleaned['Attrition'] == 'No'][var])
    stat_2.append([var,U_stat,p_value1])
```

```python
stat_2_df = pd.DataFrame(stat_2, columns =["Variable","U_stat","P_value"])
```

```python
stat_2_df.sort_values("P_value", ascending = True, inplace = True)
```

```python
stat_2_df
```

```python
stat_3 = []
```

```python
for var in numerical_feat:
    F_stat, p_value2 = stats.f_oneway(df_cleaned[df_cleaned['Attrition'] == 'Yes'][var], df_cleaned[df_cleaned['Attrition'] == 'No'][var])
    stat_3.append([var,F_stat,p_value2])
```

```python
stat_3_df = pd.DataFrame(stat_3, columns =["Variable","F_stat","P_value"])
```

```python
stat_3_df.sort_values("P_value", ascending = True, inplace = True)
```

```python
stat_3_df
```

# Modelling

```python
Est_knn = KNeighborsClassifier()
Est_Dtree = DecisionTreeClassifier()
Est_LogR = LogisticRegression()
Est_svc = SVC()
Est_RFor = RandomForestClassifier()
Est_ABoost = AdaBoostClassifier()
```

```python
knn_dict ={
    "Name": "KNeighborsClassifier",
    "Estimateur": Est_knn,
    "Paramgrid": {"n_neighbors":[2,3,4,5,6,7]}
}
```

```python
Dtree_dict = {
    "Name": "DecisionTreeclassifier",
    "Estimateur": Est_Dtree,
    "Paramgrid": {"criterion": ['entropy','gini'], "max_depth": [None,3,4,5,6,7,8,9,10]}
}
```

```python
LogR_dict = {
    "Name": "LogisticRegression",
    "Estimateur": Est_LogR,
    "Paramgrid": {"solver" : ['liblinear'], "penalty" : ['l2'], "C" : [0.001,0.01,0.1,1.0,10]}
}
```

```python
SVC_dict = {
    "Name": "SVC",
    "Estimateur": Est_svc,
    "Paramgrid": {"C": [0.001,0.01,0.1,1,10], "kernel": ["rbf","linear"], "gamma": [0.01,0.1,1]}
}
```

```python
RFor_dict = {
    "Name": "RandomForestClassifier",
    "Estimateur": Est_RFor,
    "Paramgrid": {"criterion": ["entropy","gini"], "max_depth": [None,3,4,5,6,7,8,9,10], "n_estimators": [50,100,150,200]}
}
```

```python
ABoost_dict = {
    "Name": "AdaBoostClassifier",
    "Estimateur": Est_ABoost,
    "Paramgrid": {"algorithm": ["SAMME", "SAMME.R"], "n_estimators": [50,100,150,200]}
}
```

```python
list_dict = [knn_dict,Dtree_dict,LogR_dict,SVC_dict,RFor_dict,ABoost_dict]
```

```python
Grid = {}
```

```python
X_train_np = np.ascontiguousarray(X_train.to_numpy())
y_train_np = np.ascontiguousarray(Y_train.to_numpy())
X_test_np = np.ascontiguousarray(X_test.to_numpy())
y_test_np = np.ascontiguousarray(Y_test.to_numpy())
```

```python
for val in list_dict:
    Grid[val["Name"]] = GridSearchCV(val["Estimateur"],val["Paramgrid"],cv=5)
    Grid[val["Name"]].fit(X_train_np,np.ravel(y_train_np))
    
```

```python
Grid
```

# Model selection and evaluation

```python
df_score = []
```

```python
for x in list_dict:
    scor = Grid[x["Name"]].best_estimator_.score(X_test_np,np.ravel(y_test_np))
    df_score.append([x["Name"],scor])
    
```

```python
df_score_df =pd.DataFrame(df_score, columns = ["Estimateur","Score"])
df_score_df.sort_values("Score", ascending = False, inplace = True)
```

```python
df_score_df
```

```python
df_score_df.iloc[-3:,0]
```

```python
for val in df_score_df.iloc[-3:,0]:
    del Grid[val]
```

```python
Grid
```

```python
for cle, valeur in Grid.items():
    accuracy = accuracy_score(y_test_np,valeur.best_estimator_.predict(X_test_np))
    print(f"{cle} : {accuracy}")
```

```python
for cle, valeur in Grid.items():
    predictions = valeur.best_estimator_.predict(X_test_np)
    cm = confusion_matrix(y_test_np,predictions,labels=valeur.best_estimator_.classes_)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                              display_labels=valeur.best_estimator_.classes_)
    disp.plot()
    plt.title(cle)
```

```python
for cle, valeur in Grid.items():
    print("{} \n{}\n".format(cle,classification_report(y_test_np,valeur.best_estimator_.predict(X_test_np),target_names=["Will not leave","Will leave"])))
    
```

```python
for key,val in Grid.items():
    print("{} : {} \n".format(key,val.best_params_))
```
